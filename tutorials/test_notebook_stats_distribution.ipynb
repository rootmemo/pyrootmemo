```python
import numpy as np
from pyrootmemo.stats import GammaDistribution, GumbelDistribution, WeibullDistribution, PowerDistribution
from pyrootmemo.helpers import units
```

```python
n = 100
np.random.seed(0)
xn = 2 + 8 * np.random.rand(n)
x = xn * units('m')
```

```python
fit = GammaDistribution(x)
print(fit.scale)
print(fit.shape)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = GumbelDistribution(x)
print(fit.scale)
print(fit.location)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = WeibullDistribution(x)
print(fit.shape)
print(fit.scale)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = PowerDistribution(x)
print(fit.x0)
print(fit.calc_multiplier())
print(fit.exponent)
print(fit.lower)
print(fit.upper)
fit.plot(show_fit = True)
```
