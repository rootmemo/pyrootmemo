```python
import numpy as np
from pyrootmemo.fit import calc_loglikelihood_power_class
```

```python
exponent = -0.5
xmin = 0.5
xmax = 10.0
xlower = np.array([2, 4, 6])
xupper = np.array([4, 6, 8])
weights = np.array([1, 1, 1])

par = np.array([exponent, xmin, xmax])
L = calc_loglikelihood_power_class(par, xlower, xupper, weights, 0.0)
print(L)

eps = 1e-4
dL_dpar = np.zeros(3)
for i in range(3):
    par_new = par.copy()
    par[i] = par[i] + eps
    Lnew = calc_loglikelihood_power_class(par_new, xlower, xupper, weights, 0.0)
    dL_dpar[i] = (Lnew - L) / eps
dL = calc_loglikelihood_power_class(par_new, xlower, xupper, weights, 0.0, return_jacobian = True)[1]

print(dL_dpar)
print(dL)

```
