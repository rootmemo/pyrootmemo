````python
from pyrootmemo import Parameter
from pyrootmemo.materials import SingleRoot, MultipleRoots
from pyrootmemo.materials import Soil
````

## `Parameter` - Physical Quantity with Unit

`Parameter` is a named tuple that pairs a numeric value with its unit string. It is used to construct material parameters in `pyrootmemo`

```python
from pyrootmemo import Parameter

diameter = Parameter(1.0, "mm")
```

## `SingleRoot` - A single plant root with its parameters

We are going to create a single Barley root with a given elastic modulus and diameter.

````python
barley = SingleRoot(species="hordeum_vulgare",
                           elastic_modulus=Parameter(1e9, "MPa"), 
                           diameter=Parameter(0.01, "mm"))
print(barley.diameter)
````

We can access the magnitude and the unit of each attribute

````python
print(barley.diameter.magnitude)
print(barley.diameter.units)
````

We can also see a list of attributes of a single root object

````python
print(barley.__dict__)
````

## `MultipleRoots` - An object containing multiple roots


We are first going to create 5 alder roots of different diameters, and then 2 roots of same diameter of alder and barley.

````python
alder = MultipleRoots(species="alnus_incana",
                      diameter=Parameter([1, 2, 3, 4, 5], "mm"))
print(alder.circumference)
````

## `Soil` - A soil object with its parameters

The `Soil` class stores geotechnical properties of a soil layer. Supported parameters are:
- `cohesion` (kPa)
- `friction_angle` (degrees)
- `unit_weight_bulk` (kN/m³)
- `unit_weight_dry` (kN/m³)
- `unit_weight_saturated` (kN/m³)
- `water_content` (%)

You can also supply optional `uscs_classification` and `usda_classification` strings.

Not all parameters need to be provided — only those relevant to your analysis.

### Example: Loess soil from Operstein & Frydman (2000)

The values below come from direct shear tests on a natural loess deposit reported in:

> Operstein, V. & Frydman, S. (2000). *The influence of vegetation on soil strength.*
> **Ground Improvement**, 4(2), 81–89. https://doi.org/10.1680/grim.2000.4.2.81

The bare (unvegetated) loess specimen properties used in their experiments:

| Parameter | Value | Unit |
|---|---|---|
| Cohesion | 10 | kPa |
| Friction angle | 29 | ° |
| Dry unit weight | 13.0 | kN/m³ |
| Bulk unit weight | 14.8 | kN/m³ |
| Water content | 14 | % |

USCS classification: **ML** (silt of low plasticity); USDA classification: **silt loam**.

````python
loess = Soil(
    name="loess",
    uscs_classification="ML",
    usda_classification="silt_loam",
    cohesion=Parameter(10.0, "kPa"),
    friction_angle=Parameter(29.0, "degrees"),
    unit_weight_dry=Parameter(13.0, "kN/m^3"),
    unit_weight_bulk=Parameter(14.8, "kN/m^3"),
    water_content=Parameter(14.0, "percent"),
)
print(loess.name)
print(loess.uscs_classification)
print(loess.cohesion)
print(loess.friction_angle)
````

Each attribute stores a value with its unit. You can access the magnitude and the unit separately, just like with root parameters.

````python
print(loess.cohesion.magnitude)
print(loess.cohesion.units)
print(loess.unit_weight_dry.magnitude)
print(loess.water_content.magnitude)
````
