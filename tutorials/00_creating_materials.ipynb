## `Parameter` - Physical Quantity with Unit

`Parameter` is a named tuple that pairs a numeric value with its unit string. It is used to construct material parameters in `pyrootmemo`

```python
from pyrootmemo import Parameter

diameter = Parameter(1.0, "mm")
```

## `SingleRoot` - A single plant root with its parameters

We are going to create a single Barley root with a given elastic modulus and diameter.

````python
barley = SingleRoot(species="hordeum_vulgare",
                           elastic_modulus=Parameter(1e9, "MPa"), 
                           diameter=Parameter(0.01, "mm"))
print(barley.diameter)
````

We can access the magnitude and the unit of each attribute

````python
print(barley.diameter.magnitude)
print(barley.diameter.units)
````
