```python
from pyrootmemo import Parameter
from pyrootmemo.materials import MultipleRoots
from pyrootmemo.geometry import FailureSurface
from pyrootmemo.models import Wwm
```

# Wu/Waldron model (WWM)

The Wu/Waldron model is one of the simplest root reinforcement models, 
assuming that all roots break simultaneously in tension.

## Create a model object

Define some roots, using the `MultipleRoots` class. The WWM model requires the 
following root attributes to be defined:

* Root diameter: (`diameter`)
* Root tensile strength (`tensile strength`)

```python
roots = MultipleRoots(
    species = 'test_species',
    diameter = Parameter([1, 2, 3, 4], 'mm'),
    tensile_strength = Parameter([500, 300, 100, 50], 'MPa')
    )
```

Next, create an instance of the WWM model class 

```python
wwm = Wwm(roots)
```

## Calculations

Various calculations can be made. All results are stored in dictionary object 
in the `output` attribute of the class.

### Peak force (`peak_force`)

To calculate the total maximum tensile force the roots can carry, first do the
calculation and then request the results from the `output` attribute:

```python
wwm.calc_peak_force()
print(wwm.output['peak_force'])
```

Calculation output can easily be converted to different units. For example,, 
to convert the calculated peak force to kilonewtons:

```python
print(wwm.output['peak_force'].to('kN'))
```

### Peak reinforcement (`peak_reinforcement`)

To calculate the peak reinforcement, the cross-sectional area of the failure
surface is required. A failure surface can be defined using the 
`FailureSurface` class:  

```python
failure_surface = FailureSurface(cross_sectional_area = Parameter(2, 'm^2'))
```

Secondly, the Wu/Wadron $k'$ factor should be defined, which depends on the 
orientation of the roots at failure and the angle of internal friction of the 
soil. By default, this set to $k'=1.2$:

```python
wwm.calc_peak_reinforcement(failure_surface)
print(wwm.output['peak_reinforcement'])
print(wwm.output['peak_reinforcement'].to('kPa'))
```

Note that units are all correctly handled, despite root diameters being 
defined in millimetres and the cross-sectional area of the failure surface in
meters (squared)!

$k'$ can also be manually defined, for example $k'=1$:

```python
wwm.calc_peak_reinforcement(failure_surface, k = 1.0)
print(wwm.output['peak_reinforcement'].to('kPa'))
```
