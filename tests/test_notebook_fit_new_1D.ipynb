```python
import numpy as np
from pyrootmemo.fit import GammaFit, GumbelFit, WeibullFit, PowerFit
from pyrootmemo.helpers import units
```

```python
n = 100
xn = 2 + 8 * np.random.rand(n) 
x = xn * units('m')
```

```python
fit = GammaFit(x)
print(fit.scale)
print(fit.shape)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = GumbelFit(x)
print(fit.scale)
print(fit.location)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = WeibullFit(x)
print(fit.shape)
print(fit.scale)
print(fit.calc_density(8 * units('m')))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```

```python
fit = PowerFit(x)
print(fit.x0)
print(fit.multiplier)
print(fit.exponent)
print(fit.lower)
print(fit.upper)
xx = np.array([8]) * units('m')
print(fit.calc_density(xx))
print(fit.calc_density(xx, cumulative = True))
print(fit.calc_loglikelihood(deriv = 0))
print(fit.calc_loglikelihood(deriv = 1))
print(fit.calc_loglikelihood(deriv = 2))
fit.plot()
```
