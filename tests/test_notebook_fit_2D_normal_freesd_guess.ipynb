```python
import numpy as np
from pyrootmemo.fit import LinearFit
import matplotlib.pyplot as plt
```

```python
import numpy as np
# np.random.seed(0)

x = np.linspace(1.0, 10.0, 2000)
y = 20 * x**(1.0) + np.random.normal(0.0, 3.0, len(x)) * x**(1.0)
plt.plot(x, y, 'x')
```

```python
from scipy.special import erfinv
ft = LinearFit(np.log(x), np.log(y))
beta = ft.gradient
order = np.argsort(y / x**beta)
z = (y / x**beta)[order]
q = (2 * np.arange(len(z)) + 1) / (2 * len(z))
yf = np.log(np.abs(z - np.mean(z)) / (erfinv(q) * np.sqrt(2)))
plt.plot(np.log(x[order]), yf, 'x')
ft = LinearFit(np.log(x[order]), yf)
print(ft.gradient, ft.gradient + beta)
```
