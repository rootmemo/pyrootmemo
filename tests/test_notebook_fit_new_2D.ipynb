```python
import numpy as np
from pyrootmemo.fit import PowerlawFit, LinearFit
from pyrootmemo.helpers import Parameter
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
xn = np.linspace(1.0, 10.0, 30)
yn = 10.0 * xn**(-0.5) + 5 * np.random.rand(len(xn))
x = Parameter(xn, 'mm')
y = Parameter(yn, 'MPa')
```

```python
fit = PowerlawFit(x, y, model = 'gamma')
print(fit.multiplier)
print(fit.exponent)
print(fit.shape)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'gumbel')
print(fit.multiplier)
print(fit.exponent)
print(fit.scale0)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'logistic')
print(fit.multiplier)
print(fit.exponent)
print(fit.scale0)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'weibull')
print(fit.zero_variance)
print(fit.multiplier)
print(fit.exponent)
print(fit.shape)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'normal')
print(fit.multiplier)
print(fit.exponent)
print(fit.sd_multiplier)
print(fit.sd_exponent)
```

```python
fit = PowerlawFit(x, y, model = 'normal_force')
print(fit.multiplier)
print(fit.exponent)
print(fit.sd_multiplier)
print(fit.sd_exponent)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'normal_scaled')
print(fit.multiplier)
print(fit.exponent)
print(fit.sd_multiplier)
print(fit.sd_exponent)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'normal_freesd')
print(fit.multiplier)
print(fit.exponent)
print(fit.sd_multiplier)
print(fit.sd_exponent)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'uniform')
print(fit.multiplier)
print(fit.exponent)
print(fit.width)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'uniform', algorithm = 'root')
print(fit.multiplier)
print(fit.exponent)
print(fit.width)
print(fit.calc_ks())
```

```python
fit = PowerlawFit(x, y, model = 'normal')
fit.plot(show_confidence = True, show_prediction = True, show_legend = True)
```

```python
fit_lm = LinearFit(x, y)
print(fit_lm.intercept)
print(fit_lm.gradient)
fit_lm.plot(show_confidence = True, show_prediction = True, show_legend = True)
```
